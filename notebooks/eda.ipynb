### Users reading multiple books


```python
ratings['user_id'].value_counts()
```

```python
#Users reading multiple books beyound 200 books

x = ratings['user_id'].value_counts() > 200

x[x].shape
```

```python
y = x[x].index
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)]

ratings.tail()
```

```python
# Joining ratings with books and users
ratings_with_books = ratings.merge(books, on='isbn')

ratings_with_books.head(5)
```

```python
#titles with multiple ratings
num_of_ratings = ratings_with_books.groupby('title')['ratings'].count().reset_index()

num_of_ratings.rename(columns={'ratings': 'num_of_ratings'}, inplace=True)
```

```python
num_of_ratings.head()   
```

```python
final_ratings = ratings_with_books.merge(num_of_ratings, on='title')

final_ratings.head()
```

```python
# Books which have more than 50 ratings by users

final_ratings = final_ratings[final_ratings['num_of_ratings'] > 50]

final_ratings.head(5)
```

```python
# final_ratings.drop_duplicates(['user_id', 'title'], inplace=True)

# final_ratings.head()
```

```python
final_ratings.isnull().sum()    
```

```python
# Create pivotal table for user-item matrix
book_pivot = final_ratings.pivot_table(columns='user_id', index='title', values='ratings')

book_pivot.head()
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)

book_pivot.head()
```

```python
book_pivot.shape
```

```python
# Converting the pivot table into CSR metrix

#CSR matrix (Compressed Sparse Row matrix) is an efficient way to represent sparse 
# data (where most values are zero).


from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
book_sparse
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
book_pivot.iloc[450,:]
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[355,:].values.reshape(1, -1), n_neighbors=6)

distance
```

```python
suggestion
```

```python
#Get the titles of the suggested books

for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
# Saving the book names

book_names = book_pivot.index

book_names
```

```python
book_names[6]
```

```python
# Get all suggested book names

book_name =[]
for book_id in suggestion:
    book_name.append(book_names[book_id])       

book_name
```

```python
# Exracting the image URL for the suggested books

ids_index = []

for name in book_name[0]:
    ids = np.where(final_ratings['title'] == name)[0][0]
    ids_index.append(ids)

ids_index
```

```python
# With respect to the indexes, we can extract the image URLs from the final_ratings DataFrame
for idx in ids_index:
    url = final_ratings.iloc[idx]['image_url']
    print(url)
   
```

```python
import pickle
pickle.dump(model,open('artifacts/model.pkl','wb'))
pickle.dump(book_names,open('artifacts/book_names.pkl','wb'))
pickle.dump(final_ratings,open('artifacts/final_rating.pkl','wb'))
pickle.dump(book_pivot,open('artifacts/book_pivot.pkl','wb'))
```
